```python
from langchain_community.llms import Ollama

llm = Ollama(model="llama2")

print("ðŸ¤– Chatbot using Llama 2 (via LangChain)")
print("Type 'quit' to exit.\n")

while True:
    user_input = input("You: ")
    if user_input.lower() in ["quit", "exit"]:
        break

    try:
        response = llm.invoke(f"You are a helpful company policy assistant. {user_input}")
    except ValueError as e:
        err_str = str(e)
        if "out of memory" in err_str.lower() or "cuda" in err_str.lower() or "llama runner process has terminated" in err_str.lower():
            print("Warning: Ollama model failed due to GPU/runner error (likely out of GPU memory).")
            print("Suggestions:")
            print("- Use a smaller model (e.g. a 7B variant) or run Ollama with CPU support.")
            print("- Reduce batch size / context length if configurable.")
            response = "Sorry â€” the local Llama model ran out of GPU memory. Try a smaller model or run Ollama on CPU."
        else:
            raise
    print("Bot:", response)
```
